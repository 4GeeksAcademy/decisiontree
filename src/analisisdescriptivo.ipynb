```python
import pandas as pd

total_data = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/decision-tree-project-tutorial/main/diabetes.csv", sep = ",")
total_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Obtener estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(total_data.describe())

# Información general sobre el DataFrame
print("\nInformación general:")
print(total_data.info())

# Conteo de valores únicos en cada columna
print("\nConteo de valores únicos:")
print(total_data.nunique())

# Visualizar la distribución de una variable (por ejemplo, 'variable_interes')
plt.figure(figsize=(10, 6))
plt.hist(total_data['Outcome'], bins=20, color='blue', alpha=0.7)
plt.title('Distribución de variable_interes')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```
